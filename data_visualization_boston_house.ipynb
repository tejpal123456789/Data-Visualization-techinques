Distribution plot

```
row=2
col=7
fig,ax=plt.subplots(nrows=row,ncols=col,figsize=(16,4))
cols=data1.columns
index=0
for i in range(row):
    for j in range(col):
        sns.distplot(data1[cols[index]],ax=ax[i][j],kde=True)
        index=index+1

plt.tight_layout()        
```

Scatter Plots

```
row=2
col=7
fig,ax=plt.subplots(nrows=row,ncols=col,figsize=(16,4))
cols=data1.columns
index=0
for i in range(row):
    for j in range(col):
        sns.scatterplot(x=data1[cols[index]],y=data1['price'],ax=ax[i][j])
        index=index+1

plt.tight_layout()       
```

CORELATED FEATURES AND FEATURE SELECTION


```
corr.index
```

```
def get_corelated_data(corrdata,threshold):
    feature=[]
    value=[]
    for i,index in enumerate(corrdata.index):
        if abs(corrdata[index]) > threshold:
            feature.append(index)
            value.append(corrdata[index])
    df=pd.DataFrame(data=value,index=feature,columns=['corr value'])

    return df
corr_df=get_corelated_data(corr['price'],0.5) 
corr_df           
```

Heatmap and pairplot of corealated data

```
corealted_data=data1[corr_df.index]
sns.pairplot(corealted_data)
plt.show()
```

```
sns.heatmap(corealted_data.corr(),annot=True)
```
