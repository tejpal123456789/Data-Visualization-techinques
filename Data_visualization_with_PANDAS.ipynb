```
import seaborn as sns
iris=sns.load_dataset('iris')
iris.head()
```

```
iris.plot(figsize=(15,10))
```

```
# more on line plot
x=iris.drop(labels=['sepal_width','petal_width'],axis=1)
y=iris[['sepal_width','petal_width']]
x.plot()
```

```
ax=x.plot()
y.plot(figsize=(16,10),secondary_y=True,ax=ax)
```

```
# Bar plot
df=iris.drop(['species'],axis=1)
df.iloc[0].plot(kind='bar')
```

```
df.plot.bar(figsize=(20,20))
```

```
# histogram
titanic=sns.load_dataset('titanic')
titanic['pclass'].plot.hist()
```

```
titanic['pclass'].unique()
```

```
titanic['survived'].plot.hist()
```

```
# stacked plot
df1=pd.DataFrame(np.random.randn(10,4),columns=['a','b','c','d'])
df1.head()
```

```
df1.plot.bar(stacked=True)
```

```
# histograms
iris.plot.hist()
```

```
iris.plot.hist(bins=50,stacked=True)
```

```
from matplotlib import pyplot as plt
df2=iris.drop(['species'],axis=1)
df2.hist(color='r',bins=30)
plt.show()
```

```
# Box plot
df2.head()
```

```
df2.plot.box(vert=False)
```

```
# area and scatter plot
df.plot.area()
```

```
df.plot.area(stacked=False)
```

```
# scatter
df2.plot.scatter(x='sepal_width',y='petal_width')
```

```
# hex and pie plot
df2.head()
```

```
df2.plot.hexbin(x='sepal_length',y='petal_length',C='sepal_width',gridsize=10)
```

```
# pie plot
df2.iloc[0].plot.pie(autopct='%.2f')
```

```
# scatter matrix and sub plot
from pandas.plotting import  scatter_matrix
scatter_matrix(df2,figsize=(10,10),diagonal='kde',color='r')
plt.show()
```

```
# subplots
df2.plot(subplots=True,sharex=False,figsize=(10,10))
```

```

df2.plot(subplots=True,sharex=False,figsize=(10,10),layout=(2,2))
```
